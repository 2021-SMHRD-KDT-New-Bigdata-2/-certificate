```python
url1 = "https://edu.goorm.io/category/programming"
driver = wb.Chrome()
driver.get(url1)
time.sleep(1)

for i in range(1,21):
    try:
        #각 항목마다 들어가기
        btn = driver.find_element_by_css_selector('div.Aj2j_L > div:nth-child('+str(i)+')>a')
        btn.click()
        time.sleep(3)
        #카테고리(cate)
        cate = soup.select('div:nth-child(2) > div._2yM5um')
        for j in range(len(cate)):
            cate_list.append(cate[j].text)
        try:
            #수강후기(review)
            review = soup.select('p._1KbNtk.card-text')
            #각 리스트의 텍스트만 다른 리스트에 저장
            for i in range(len(review)):
                review_list.append(review[i].text)
        except:
            review_list.append("리뷰없음")
        driver.back()
        time.sleep(2)
    except:
        body = driver.find_element_by_css_selector('body') 
        body.send_keys(Keys.PAGE_DOWN) # 페이지 다운
        time.sleep(1) # 1초 딜레이
        continue
```

```python
#첫번째 페이지, 팝업창
class101_num_list = []
class101_review_list = []
review_list = []

url1 = "https://class101.net/search?category=604f1c9756c3676f1ed00397"
driver = wb.Chrome()
driver.get(url1)
time.sleep(8)
btn = driver.find_element_by_css_selector('ul > li:nth-child(1) > a')
btn.click()
time.sleep(8)
btnclose = driver.find_element_by_css_selector('.ab-close-button')
btnclose.click()
soup = bs(driver.page_source,"lxml")
num= soup.select_one('dd.KlassSummarySection__DefinitionDescription-lcwqnj-4.jeiEOD')
class101_num_list.append(num.text)
try:
    btnclose = driver.find_element_by_css_selector('.ab-close-button')
    btnclose.click()
    time.sleep(3)
    btn2 = driver.find_element_by_css_selector('.LinkBlock-ky5rdu-0.ButtonBase__LinkButton-sc-1lxfn99-3.gXokuE.iFkzVH')
    btn2.click()
    time.eep(5)
    soup = bs(driver.page_source,"lxml")
    for i in range(1,4):
        review = soup.select('div.KlassStoryViewController__PostListContainer-sc-1cno7v-2.iqMMGN > div > div:nth-child('+str(i)+') > div > a > div > div.Reply__ContentWrapper-sc-9g4w2s-3.eMaRXm > div > pre > div')
        for k in range(len(review)):
            review_list.append(review[k].text)
    class101_review_list.append(review_list)
    driver.back()
    time.sleep(1)
    driver.back()
    time.sleep(1)
except:
    print("팝업창 안떠서 안됨")
```

### 25번째에서 중단되는 문제 해결을 위한것

```python
for i in tqdm(range(1,7)):
    btn = driver.find_element_by_css_selector('div.GridList__Container-sc-1qd26lc-0.fIyoAK > ul > li:nth-child('+str(i)+') > a')
    btn.click()
    time.sleep(10)
    #강의길이
    soup = bs(driver.page_source,"lxml")
    num = soup.select_one('.KlassSummarySection__DefinitionDescription-lcwqnj-4.jeiEOD')
    class101_num_list.append(num.text)
    time.sleep(10)
    try:
        soup = bs(driver.page_source,"lxml")
        for j in range(1,4):
            review = soup.select('.ReviewV2Card__Content-sc-1mma0rc-10.jQJcjX')
            for k in range(len(review)):
                    review_list.append(review[k].text)
        class101_review_list.append(review_list)
        driver.back()
        time.sleep(1)
    except:
        btn2 = driver.find_element_by_css_selector('#product-view-review > section > div > a:nth-child(8)')
        btn2.click()
        time.sleep(5)
        soup = bs(driver.page_source,"lxml")
        for i in range(1,4):
            review = soup.select('.Timer__Day-sc-1mgbdlp-1.bxcQif')
        for k in range(len(review)):
            review_list.append(review[k].text)
        class101_review_list.append(review_list)
        driver.back()
        time.sleep(1)
```
