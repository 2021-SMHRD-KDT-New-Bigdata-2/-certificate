```python
import pandas as pd
```

```python
gooreum = pd.read_csv("구름edu_전처리_완료.csv", index_col=0)
gooreum
```

```python
def replace_bs(x):
    st_ind = x.find('[')
    ed_ind = x.find(']')
    return x[st_ind+1:ed_ind].replace("\\n"," ").replace("\'",'"').replace("\",","\"DELIMITER111")
```

```python
gooreum["리뷰"] = gooreum["리뷰"].apply(replace_bs)
```

```python
gooreum.to_csv("구름edu_전처리_찐완료.csv", encoding="utf-8-sig")
```
